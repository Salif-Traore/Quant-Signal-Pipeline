```python
import os
import pandas as pd
import matplotlib.pyplot as plt

# --- Config ---
ticker = "^VIX"
file_path = f"data/prepared/{ticker}_with_features.parquet"

# Check file existence
if not os.path.exists(file_path):
    raise FileNotFoundError(f"File not found: {file_path}. Make sure your working directory is the project root.")

# Load data
df = pd.read_parquet(file_path)

# Show first few rows
print(f"First 10 rows of {ticker} data:")
display(df.head(10))

# Plot
fig, ax1 = plt.subplots(figsize=(14,7))

ax1.plot(df.index, df['Close'], label='Close Price', color='blue')
ax1.set_xlabel('Date')
ax1.set_ylabel('Close Price', color='blue')
ax1.tick_params(axis='y', labelcolor='blue')

ax2 = ax1.twinx()
ax2.plot(df.index, df['signal_12m_return'], label='12m Return Signal', color='red', alpha=0.6)
ax2.set_ylabel('12m Return Signal', color='red')
ax2.tick_params(axis='y', labelcolor='red')

plt.title(f"{ticker} Close Price and 12-Month Return Signal")
fig.tight_layout()
plt.show()
```
